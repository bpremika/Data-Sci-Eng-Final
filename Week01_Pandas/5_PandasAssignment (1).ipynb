## How many rows are there in the GBvideos.csv after removing duplications?

```python
gb_df = pd.read_csv('data/GBvideos.csv')

gb_df.drop_duplicates(inplace=True)

gb_df
```

```python
gb_df.shape[0] #ANS
```

## How many VDO's that contain at least one record (row) of "dislikes" more than "likes"?  <font color=red>DO NOT group by the data and make sure that you count only unique title!</font>

```python
condition_list = gb_df.dislikes > gb_df.likes

gb_df[condition_list].title.nunique() #ANS
```

## How many VDO that are trending on 22 Jan 2018 with comments more than 10,000 comments?

```python
condition_list = (gb_df.trending_date == '18.22.01') & (gb_df.comment_count > 10000)

gb_df[condition_list].shape[0]
```

## Which date that has the minimum average number of comments per VDO?

```python
date_w_avg_comments = gb_df.groupby('trending_date').comment_count.mean().reset_index(name='avg_comments')
date_w_avg_comments[date_w_avg_comments.avg_comments == date_w_avg_comments.avg_comments.min()]
```

## Compare "Sports" and "Comady", how many days that there are more total daily views of VDO in "Sports" category than in "Comady" category?

```python
import json
with open('data/US_category_id.json') as fd:
    cat = json.load(fd)

cat_list = []
for d in cat['items']:
    cat_list.append((int(d['id']), d['snippet']['title']))

cat_df = pd.DataFrame(cat_list, columns=['id', 'category'])
cat_df
```

```python
gb_df_withcat = gb_df.merge(cat_df, left_on='category_id', right_on='id')
gb_df_withcat
```

```python
comady_list = gb_df_withcat.category == 'Comedy'
print(comady_list)
sport_list = gb_df_withcat.category == 'Sports'
print(sport_list)
```

```python
comady_df = gb_df_withcat[comady_list].groupby('trending_date').views.sum().reset_index(name='total_views_comedy')
comady_df
```

```python
sport_df = gb_df_withcat[sport_list].groupby('trending_date').views.sum().reset_index(name='total_views_sport')
```

```python
comady_sport_df =comady_df.merge(sport_df, on = 'trending_date', how = 'inner')
comady_sport_df[comady_sport_df.total_views_sport > comady_sport_df.total_views_comedy].shape[0]
```
